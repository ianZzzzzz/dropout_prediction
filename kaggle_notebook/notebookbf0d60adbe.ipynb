```python
# find file path under the folder
def return_file(path):
    path_dict = {}    
    for dirname, _, filenames in os.walk(path):    
        
        for filename in filenames:            
           
            path = os.path.join(dirname, filename)
            path_dict[filename] = path
           # print(path)
            #yield path       
    return path_dict 
            
```

```python
path = '/kaggle/input'
path_dict = return_file(path)

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```

```python
# set data type for each columns
columns_type_dict = {}

columns_type_dict['test_log.csv'] = {
      'enroll_id':'int64'
    ,  'username':'int64'
    , 'course_id':'str'
    ,'session_id':'str'
    ,    'action':'str'
    ,    'object':'str'
    ,      'time':'str'
}
columns_type_dict['train_log.csv'] = {
      'enroll_id':'int64'
    ,  'username':'int64'
    , 'course_id':'str'
    ,'session_id':'str'
    ,    'action':'str'
    ,    'object':'str'
    ,      'time':'str'
}

columns_type_dict['test_truth.csv'] ={
    'enroll_id': 'int64'
    ,   'truth':'boolean'
}

columns_type_dict['train_truth.csv'] ={
    'enroll_id': 'int64'
    ,   'truth':'boolean'
}

columns_type_dict['user_info.csv'] = {
      'user_id':'int64'
    ,  'gender': 'boolean'
    ,'educatin': 'int8'
    ,   'birth':'int8' # user's birth year
}

columns_type_dict['course_info.csv'] = {
             'id' : 'int64'       # the id (number) of course, which is used in dropout prediction dataset.
    , 'course_id' : 'str' # the id (string) of course, which is used in tracking log.
    ,     'start' : 'str'    # course start time.
    ,     'end'   : 'str'    # course end time.
    ,'course_type': 'boolean' # course study mode (0: instructor-paced course, 1: self-paced course).
    ,  'category' : 'str' # the category of course.
}
```

```python
# load raw data to dict: raw_data

raw_data = {}
for file in path_dict:
    raw_data[file[:-4]] = pd.read_csv(path_dict[file], dtype = columns_type_dict[file], index_col = 0)  
    print(
        file[:-4],"\n",
        "Rows: {:,}".format(len(raw_data[file[:-4]])), "\n" +
        "Columns: {}".format(len(raw_data[file[:-4]].columns)))
    
```
